# REVENUE

Step 1 - "unit_price_total"

```python
orderlines_qu["unit_price_total"] = orderlines_qu["product_quantity"] * orderlines_qu["unit_price"]  # this is the price per order (??)
```

Step 2 - Group by order_id, summarising by the sum of unit_price_total

```python
orderlines_qu_group_order_df = orderlines_qu.groupby("id_order", as_index=False)["unit_price_total"].sum()
```

What is the average difference between total_paid and unit_price_total?

merge our grouped orderlines table with the orders table, so we can calculate the differences in prices.

```python
diff_df = orders_qu.merge(orderlines_qu_group_order_df,
                               how = 'inner',
                               left_on="order_id",
                               right_on="id_order"
                               )
```

```python
diff_df["difference"] = diff_df["total_paid"] - diff_df["unit_price_total"]         #discount? No, is just the difference that what should be paid ( quantity by price per product) and the price paid per order
diff_df.head()
```

```python
diff_df.difference.mean().round(2)   # what is .difference(). What is that for? JH- is for the total paid and total pice, that might be because of shipping cost, but I am not sure if we can explain all the dif by shipping cost
```

```python
diff_df.difference.describe()
```

```python
# Grouping by 'id_order' lets find  out the total amount for each individual order.
# This sum represents the total price based on unit price and quantity for each order.
grouped_orderlines_qu = orderlines_qu.groupby('id_order')['unit_price_total'].sum()
```

```python
# By merging, its creating a combined dataset to directly compare and analyze the discrepancies between
# what was paid in total (from the orders_qu table) and what the total amount should have been (from our calculation).
merged_orders_qu = pd.merge(orders_qu, grouped_orderlines_qu, left_on='order_id', right_on='id_order')
```

```python
# so, to calculate the difference between the total paid and our calculated total for each order.
# This difference will help identify potential discrepancies like shipping costs or vouchers.
merged_orders_qu['price_difference'] = merged_orders_qu['total_paid'] - merged_orders_qu['unit_price_total']

# Average (mean) difference:
average_difference = merged_orders_qu['price_difference'].mean().round(2)
print(f"The average difference between total paid and unit price total is: {average_difference} euros")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Use the new .isocalendar() method
weekly_sales = mega_merge.groupby([mega_merge['date'].dt.year, mega_merge['date'].dt.isocalendar().week])['unit_price_total'].sum().reset_index()
weekly_sales.columns = ['Year', 'Week', 'Total Sales']

weekly_discount = mega_merge.groupby([mega_merge['date'].dt.year, mega_merge['date'].dt.isocalendar().week])['discount_(%)'].mean().reset_index()
weekly_discount.columns = ['Year', 'Week', 'Avg Discount']

# Create a "Year-Week" column for plotting purposes
weekly_sales['Year-Week'] = weekly_sales['Year'].astype(str) + '-W' + weekly_sales['Week'].astype(str).str.zfill(2)
weekly_discount['Year-Week'] = weekly_discount['Year'].astype(str) + '-W' + weekly_discount['Week'].astype(str).str.zfill(2)

# Merge the two dataframes on the 'Year' and 'Week' columns
merged_df = pd.merge(weekly_sales, weekly_discount, on=['Year', 'Week'])

# Plotting
fig, ax = plt.subplots(figsize=(14, 7))
fig.subplots_adjust(right=0.75)

twin1 = ax.twinx()

# Plotting the total sales
p1, = ax.plot(merged_df['Year-Week_x'], merged_df['Total Sales'], label='Revenue', color="blue")

# Plotting the average discount
p2, = twin1.plot(merged_df['Year-Week_x'], merged_df['Avg Discount'], linestyle='dashdot', label='Avg Discount (%)', color="darkgreen")

# Setting labels and title
ax.set_xlabel("Year-Week")
ax.set_ylabel("Revenue", color="blue")
ax.get_yaxis().get_major_formatter().set_scientific(False)
twin1.set_ylabel("Average Discount (%)", color="darkgreen")

# Setting tick colors to match line colors
ax.tick_params(axis='y', colors="blue")
twin1.tick_params(axis='y', colors="darkgreen")

# Rotating x-tick labels for better clarity
ax.set_xticks(merged_df['Year-Week_x'])
ax.set_xticklabels(merged_df['Year-Week_x'], rotation=45, ha='right')

ax.set_title("Revenue and Average Discount Over Time (Weekly)")

# Displaying the legend
ax.legend(handles=[p1, p2])

plt.tight_layout()
plt.show()
```
